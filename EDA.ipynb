```python
# Fill missing numeric values with the median
data_copy['calories'] = data_copy['calories'].fillna(data_copy['calories'].median())
data_copy['protein'] = data_copy['protein'].fillna(data_copy['protein'].median())
data_copy['fat'] = data_copy['fat'].fillna(data_copy['fat'].median())
data_copy['sodium'] = data_copy['sodium'].fillna(data_copy['sodium'].median())
```

```python
data_copy.isnull().sum()
```

```python
duplicates = data_copy[data_copy.duplicated()]
duplicates.head()
```

```python
# Keep only the first occurrence of each duplicate
data_cleaned = data_copy.drop_duplicates(keep='first')

# Confirm the number of remaining duplicates
remaining_duplicates = data_cleaned.duplicated().sum()
print(f"Remaining duplicates after keeping first: {remaining_duplicates}")
```

```python
condition_ratings = data_cleaned['rating'] != 0

# Condition 2: Calories of 0
condition_calories = data_cleaned['calories'] != 0

# Condition 3: Recipes with fat, protein, and sodium all equal to 0
condition_nutrients = ~((data_cleaned['fat'] == 0) &
                        (data_cleaned['protein'] == 0) &
                        (data_cleaned['sodium'] == 0))

# Combine all conditions using bitwise AND
valid_conditions = condition_ratings & condition_calories & condition_nutrients

# Filter the data
data_cleaned = data_cleaned[valid_conditions]
```

```python
def remove_outlier_IQR(df1,df_column):
    Q1=df1[df_column].quantile(0.25)
    Q3=df1[df_column].quantile(0.75)
    IQR=Q3-Q1
    vysledek=df1[~((df1[df_column] < (Q1 - 1.5 * IQR)) |(df1[df_column] > (Q3 + 1.5 * IQR)))]
    return vysledek
```

```python
df_final = remove_outlier_IQR(df,"calories")
df_final = remove_outlier_IQR(df_final,"sodium")
```

```python
import numpy as np
from scipy import stats
z_scores_fat = np.abs(stats.zscore(data_cleaned['fat']))

# Define a threshold for identifying outliers (typically Z-score > 3)
threshold = 3

# Identify outliers (rows where the Z-score is greater than the threshold)
outliers_fat = np.where(z_scores_fat > threshold)

# Display the rows that are identified as outliers
print("Outliers in Fat Column:\n", data_cleaned.iloc[outliers_fat])

# Remove outliers (keep rows where the Z-score is less than the threshold)
data_no_outliers_fat = data_cleaned[z_scores_fat < threshold]

# Check the shape of the dataset before and after removing outliers
print(f"Original data shape: {data_cleaned.shape}")
print(f"Data shape after removing fat outliers: {data_no_outliers_fat.shape}")
```

```python
def recommend_dishes(meal_type, num_recommendations=3):
    # Filter for the specified meal type
    if meal_type == 'breakfast':
        meal_column = 'breakfast'
    elif meal_type == 'lunch':
        meal_column = 'lunch'
    elif meal_type == 'dinner':
        meal_column = 'dinner'
    else:
        print("Invalid meal type specified.")
        return

    # Filter the dataset for the specified meal type (assuming 1 indicates the dish is part of that meal)
    meal_recipes = df[df[meal_column] == 1]  # Adjust this if the values are not binary

    # Sort by rating (descending) and calories (ascending)
    meal_recipes_sorted = meal_recipes.sort_values(by=['rating', 'calories'], ascending=[False, True])

    # Select top dishes
    top_dishes = meal_recipes_sorted[['title', 'rating', 'calories']].head(num_recommendations)

    return top_dishes

# Recommend dishes for each meal type
breakfast_recommendations = recommend_dishes('breakfast')
lunch_recommendations = recommend_dishes('lunch')
dinner_recommendations = recommend_dishes('dinner')

# Display the recommendations
print("Breakfast Recommendations:\n", breakfast_recommendations)
print("\nLunch Recommendations:\n", lunch_recommendations)
print("\nDinner Recommendations:\n", dinner_recommendations)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of occasion/event tags to analyze
occasion_tags = [
    'birthday', 'anniversary', 'backyard bbq', 'graduation',
    'thanksgiving', 'easter', 'halloween', 'valentine\'s day',
    'christmas', 'new year\'s day'
]

# Initialize a dictionary to store the count of dishes for each occasion
occasion_counts = {}

# Count the number of dishes for each occasion/event tag
for tag in occasion_tags:
    if tag in df.columns:
        occasion_counts[tag] = df[df[tag] == True].shape[0]

# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(occasion_counts.keys(), occasion_counts.values(), color='orange')
plt.title('Number of Dishes by Occasion/Event Tags', fontsize=14)
plt.xlabel('Occasion/Event Tag', fontsize=12)
plt.ylabel('Number of Dishes', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python


# Define a function to recommend dishes based on occasion
def recommend_dishes_by_occasion(df, occasion_tag):
    # Ensure the specified occasion tag exists in the columns
    if occasion_tag not in df.columns:
        print(f"Error: '{occasion_tag}' column not found in the dataset.")
        return

    # Filter the recipes for the specified occasion
    occasion_recipes = df[df[occasion_tag] == True]

    # Sort the recipes by rating (descending) to get the top-rated ones
    recommended_dishes = occasion_recipes.sort_values(by='rating', ascending=False)

    # Display the top 5 recommended dishes for the specified occasion
    print(f"Top recommended dishes for '{occasion_tag}':")
    print(recommended_dishes[['title', 'rating']].head(5))

# Example usage for different occasions
occasion_list = ['birthday', 'anniversary', 'backyard bbq', 'new year\'s day']

for occasion in occasion_list:
    recommend_dishes_by_occasion(df, occasion)
    print("\n" + "="*40 + "\n")  # Separator for better readability
```
